```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data_path = "../data/archive/data.csv"
df = pd.read_csv(data_path)
```

```python
# Display basic dataset information
print("\nDataset Overview:\n")
print(df.info())
print("\nFirst Few Rows:\n")
print(df.head())
```

```python
# Check for missing values
print("\nMissing Values:\n")
print(df.isnull().sum())
print("\nPercentage of Missing Values:\n")
print(df.isnull().sum() / len(df) * 100)
```

```python
# Check for duplicate rows
print("\nDuplicate Rows:\n")
print(df.duplicated().sum())
print("\nPercentage of Duplicate Rows:\n")
print(df.duplicated().sum() / len(df) * 100)
```

```python
# Descriptive statistics
print("\nDescriptive Statistics:\n")
df.iloc[:,1:].describe().T
```

```python
# Outlier Detection using Boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplots for Outlier Detection")
plt.show()
```

```python
# Outlier Detection using IQR Method
def detect_outliers_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
    return outliers

outliers = detect_outliers_iqr(df.select_dtypes(include=[np.number]))
print("\nOutliers detected (IQR method):\n")
print(outliers)
```

```python
# Feature Distributions
plt.figure(figsize=(15, 10))
df.hist(bins=30, figsize=(15, 10), edgecolor='black')

plt.suptitle("Feature Distributions", fontsize=16)

plt.subplots_adjust(hspace=0.5, wspace=0.3)  

plt.show()
```

```python
# Correlation Analysis
plt.figure(figsize=(20, 15))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Feature Relationships (Scatter Plots)
feature_pairs = [('C', 'Fatigue'), ('Cr', 'Fatigue'), ('Si', 'Fatigue'), ('RedRatio', 'Fatigue')]
plt.figure(figsize=(12, 8))
for i, (x, y) in enumerate(feature_pairs, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=df, x=x, y=y)
    plt.title(f"{x} vs {y}")
plt.tight_layout()
plt.show()
```

```python
# This will show scatter plots between all selected features.
# The diagonal shows KDE (density) plots instead of histograms.
selected_features = ['C', 'Si', 'Cr', 'RedRatio', 'Fatigue']
sns.pairplot(df[selected_features], diag_kind='kde')
plt.show()
```

```python
# Automate Feature Selection Based on Correlation

# Define a meaningful threshold
corr_threshold = 0.3  
# Compute correlations with Fatigue Strength
correlated_features = df.corr()['Fatigue'].abs().sort_values(ascending=False) 
# Select features with correlation > 0.3
top_features = correlated_features[correlated_features > corr_threshold].index.tolist() 
# Remove the target variable itself
top_features.remove('Fatigue')
# Remove SI .No
top_features.remove('Sl. No.')

print("Top correlated features with Fatigue Strength:", top_features)

# This selects features most correlated with fatigue strength.
# We can then visualize scatter plots only for important relationships.
```

```python
# Plot scatter plots for all selected features
plt.figure(figsize=(12, 20))
for i, feature in enumerate(top_features, 1):
    plt.subplot(len(top_features) // 2 + 1, 2, i)
    sns.scatterplot(data=df, x=feature, y='Fatigue')
    plt.title(f"{feature} vs. Fatigue Strength")

plt.tight_layout()
plt.show()
```
